```python
fig, axarr = plt.subplots(4, 2, figsize=(14, 20))

# School vs G3
df.groupby('school')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[0][0], fontsize=11)
axarr[0][0].set_title("School vs Final Grade (G3)", fontsize=16)

# Sex vs G3
df.groupby('sex')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[0][1], fontsize=11)
axarr[0][1].set_title("Gender vs Final Grade (G3)", fontsize=16)

# Mother Job vs G3
df.groupby('Mjob')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[1][0], fontsize=11)
axarr[1][0].set_title("Mother Job vs Final Grade (G3)", fontsize=16)

# Father Job vs G3
df.groupby('Fjob')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[1][1], fontsize=11)
axarr[1][1].set_title("Father Job vs Final Grade (G3)", fontsize=16)

# Guardian vs G3
df.groupby('guardian')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[2][0], fontsize=11)
axarr[2][0].set_title("Guardian vs Final Grade (G3)", fontsize=16)

# Study Time vs G3
df.groupby('studytime')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[2][1], fontsize=11)
axarr[2][1].set_title("Study Time vs Final Grade (G3)", fontsize=16)

# Internet vs G3
df.groupby('internet')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[3][0], fontsize=11)
axarr[3][0].set_title("Internet Access vs Final Grade (G3)", fontsize=16)

# Mother Education vs G3
df.groupby('Medu')['G3'].mean().sort_values(ascending=False)\
  .plot.bar(ax=axarr[3][1], fontsize=11)
axarr[3][1].set_title("Mother Education vs Final Grade (G3)", fontsize=16)

plt.subplots_adjust(hspace=1.0, wspace=0.5)
sns.despine()
plt.show()
```
